```python
# Import all the needed packages:

import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

import cartopy.crs as ccrs
import cartopy
import cartopy.feature as cfeature

from geopy.geocoders import Nominatim
from geopy.distance import geodesic
```

```python
# The Southern Ocean and Antarctica 
```

```python
fig = plt.figure(figsize=(10,8))
```

```python
ax = plt.axes(projection=ccrs.SouthPolarStereo())
```

```python
# Define the axes
ax.coastlines(linewidth = 0.8)
```

```python
# Define the extent for our question
extent = [-180, 180, -90, -60]
ax.set_extent(extent, crs=ccrs.PlateCarree())

```

```python
gl = ax.gridlines(draw_labels=True, alpha = 0.4, color = 'gray')
gl.right_labels=False
gl.top_labels=False
```

```python
# Adding some features to the map
ax.add_feature(cartopy.feature.LAND, facecolor= 'lightgray')
ax.add_feature(cartopy.feature.OCEAN, facecolor="dodgerblue")
ax.add_feature(cartopy.feature.COASTLINE, facecolor = 'dimgray')
ax.text(0, -90, 'South Pole', fontsize=14, ha='center', transform=ccrs.PlateCarree())
ax.text(0, -95, 'West Antarctica', fontsize=14, ha='right', transform=ccrs.PlateCarree())
ax.text(0, -75, 'East Antarctica', fontsize=14, ha='left', transform=ccrs.PlateCarree())
```

```python
# Plotting features
plt.title("Map Projection: Antarctica and the Southern Ocean")
plt.tight_layout()
fig.savefig('Map Projection: Antarctica and the Southern Ocean.png', dpi = 300)
plt.show()
```
