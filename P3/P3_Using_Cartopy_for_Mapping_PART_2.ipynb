```python
# Import all the needed packages:

import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

import cartopy.crs as ccrs
import cartopy
import cartopy.feature as cfeature

from geopy.geocoders import Nominatim
from geopy.distance import geodesic
```

```python
#Plot the figure

fig = plt.figure(figsize=(10,8))
```

```python
# Add the projection:

ax = plt.axes(projection=ccrs.Mollweide())
```

```python
#Add the extent for the question:

extent_2 = [-60, 30, -50, -20]
ax.set_extent(extent_2, crs=ccrs.PlateCarree())
```

```python
#Add Map features:

ax.add_feature(cartopy.feature.LAND, facecolor= 'lightgray')
ax.add_feature(cartopy.feature.OCEAN, facecolor="dodgerblue")
ax.add_feature(cartopy.feature.COASTLINE, edgecolor = 'dimgray')
ax.add_feature(cartopy.feature.BORDERS, edgecolor = 'black')
```

```python
#Add coastlines

ax.coastlines(linewidth = 0.8, linestyle = '--', alpha = 0.2)
```

```python
# Add gridlines

gl = ax.gridlines(draw_labels=True)
gl.xlines = False
gl.ylines = True
gl.right_labels=False
gl.top_labels=False
```

```python
geolocator = Nominatim(user_agent = 'map')
```

```python
#List the places

places = ['Cape Town', 'Walvis Bay', 'Rio de Janeiro' , 'Montevideo']
```

```python
address = []
```

```python
# Locate the cities
for place in places:
    loc = geolocator.geocode(place, language='en')
    address.append(loc)
print(address)
```

```python
# Cape Town 

ax.plot(address[0].longitude,
        address[0].latitude,
        marker='o',
        color='red',
        markersize=6,
        transform=ccrs.PlateCarree())

ax.text(address[0].longitude+1,
        address[0].latitude,
        "Cape Town",
        fontsize=9,
        transform=ccrs.PlateCarree())
```

```python
#Walvis Bay

ax.plot(address[1].longitude,
        address[1].latitude,
        marker='o',
        color='red',
        markersize=6,
        transform=ccrs.PlateCarree())

ax.text(address[1].longitude+1,
        address[1].latitude,
        "Walvis Bay",
        fontsize=9,
        transform=ccrs.PlateCarree())
```

```python
# Rio de Janeiro

ax.plot(address[2].longitude,
        address[2].latitude,
        marker='o',
        color='red',
        markersize=6,
        transform=ccrs.PlateCarree())

ax.text(address[2].longitude+1,
        address[2].latitude,
        "Rio de Janeiro",
        fontsize=9,
        transform=ccrs.PlateCarree())
```

```python
#Montevideo

ax.plot(address[3].longitude,
        address[3].latitude,
        marker='o',
        color='red',
        markersize=6,
        transform=ccrs.PlateCarree())

ax.text(address[3].longitude+1,
        address[3].latitude,
        "Montevideo",
        fontsize=9,
        transform=ccrs.PlateCarree())
```

```python

gl = ax.gridlines(draw_labels=True, alpha = 0.5)

gl.xlabels_top = False
gl.xlabels_bottom = False

gl.ylabels_left = True
gl.ylabels_right = True
```

```python
plt.title('Coastal Cities of the South Atlantic Ocean')
plt.tight_layout()
plt.show()
```
